```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

file_NIS_laser = '../build/NIS_laser.txt';
file_NIS_radar = '../build/NIS_radar.txt';

with open(file_NIS_laser, 'r') as f:
    NIS_laser_list = f.read().splitlines()
    NIS_laser_list = np.array(NIS_laser_list)
    NIS_laser_list = NIS_laser_list.astype(np.float)
    f.close()
    
with open(file_NIS_radar, 'r') as f:
    NIS_radar_list = f.read().splitlines()
    NIS_radar_list = np.array(NIS_radar_list)
    NIS_radar_list = NIS_radar_list.astype(np.float)
    f.close()
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(NIS_laser_list, '-')
plt.ylabel('some numbers')
plt.axhline(y=5.991, color='r', linestyle='-')
plt.title('NIS - Laser')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(NIS_radar_list[1:], '-')
plt.ylabel('some numbers')
plt.axhline(y=7.815, color='r', linestyle='-')
plt.title('NIS - Radar')
plt.show()
```
